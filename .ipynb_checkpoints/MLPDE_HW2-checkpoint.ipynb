## Setup

```python
# Vectorized potential computation

# V, x vector, l lambda
def V_batch(x, l=0.03):
    # x: [B, d]
    B, d = x.shape
    h = 1 / (1 + d)

    x_prev = F.pad(x[:, :-1], (1, 0), value=0.0)  # [B, d]

    diff_term = 0.5 * l * ((x - x_prev) / h) ** 2
    well_term = (1 - x ** 2) ** 2 / (4 * l)

    return (diff_term + well_term).sum(dim=1)  # [B]
```

```python
# lagevin sampler based on pytorch auto grad
def langevin(x, f, itr, T=1.0, h=1e-3):
    """
    x: [B, d] tensor, will be modified in-place
    """
    x = x.detach()
    x.requires_grad_(True)

    for _ in range(itr):
        y = f(x)  # [B]
        grad = torch.autograd.grad(y, x, grad_outputs=torch.ones_like(y), retain_graph=False, create_graph=False)[0]
        noise = torch.randn_like(x)
        x = x - h * grad + (2 * T * h) ** 0.5 * noise
        x = x.detach()  # stop autograd graph accumulation
        x.requires_grad_(True)

    return x.detach()
```

```python
# samples N langevin based samples of dimension d
def sample(N, d, T = 8):
    # Config
    L = 2.5           # domain range
    itr = 100         # number of Langevin steps
    h = 3e-4          # step size

    # Initialize from Uniform[-L, L]^d
    x0 = (torch.rand(N, d) * 2 * L) - L  # shape [B, d]

    # Run simulation
    x_final = langevin(x0, V_batch, itr=itr, T=T, h=h)
    return x_final
```

```python
def visualize_dist(x_np):

    dims_to_plot = 8
    fig, axs = plt.subplots(dims_to_plot, dims_to_plot, figsize=(15, 15))

    for i in range(dims_to_plot):
        for j in range(dims_to_plot):
            ax = axs[i, j]

            if i == j:
                # Diagonal: nice 1D histogram with optional KDE
                sns.histplot(x_np[:, i], bins=50, kde=True, stat='density', ax=ax, color='gray')
            else:

                # Off-diagonal: fast 2D histogram heatmap
                ax.hist2d(x_np[:, j], x_np[:, i], bins=50, density=True, cmap='viridis')

            # Hide ticks for cleaner look
            if i < dims_to_plot - 1:
                ax.set_xticks([])
            if j > 0:
                ax.set_yticks([])

            if i == dims_to_plot - 1:
                ax.set_xlabel(f"$x_{j+1}$")
            else:
                ax.set_xlabel("")
            if j == 0:
                ax.set_ylabel(f"$x_{i+1}$")
            else:
                ax.set_ylabel("")

    plt.tight_layout()
    plt.show()
```

# Diffusion training

We have the following diffusion process:
$$d\overrightarrow X_t = b( X_t)dt+\sqrt{2}dB_t$$
$$d\overleftarrow X_t = (-b( X_t) +2\nabla \log \mu_{T-t} (X_t))dt+\sqrt{2}dB_t,$$
where we approximate $\nabla \log \mu_{T-t}$ by $S_t$:
$$\min_{S} ||S_t||^2_{\mu_{T-t}}+2\left<\nabla \cdot S_t, \hat \mu_{T-t}\right>.$$

In the case that we have completely discrete sample points, the above function degenerates into:
$$\frac{1}{N}\sum_n^N S((x,t)_n)^2 + 2 \nabla \cdot S((x,t)_n)$$

Note the dimension of $S_t$ is same as $\nabla \log \mu_{T-t}$, so $S_t: \mathbb{R}^d \rightarrow \mathbb{R}^d$,
$$S:\mathbb{R}^{d+1}\rightarrow \mathbb{R}^d.$$

Instead of implementing b, we will directly generate the initial data ($\overleftarrow X_0$) around the mean of samples without any loss of generality.

## training set generation

```python
# diffusion based sampler
# t1 is final diffusion time, h is step size
def diffusion_sampler(t1, h, X0, viz=False):
    d = X0.shape[1]
    N = X0.shape[0]
    buffer = []
    ts = []
    # without disturbing the original samples
    X = X0.clone()
    buffer.append(X)
    ts.append(t1)
    while(t1>0):
        # scale by sqrt(h) due to quadratic variance
        dB = torch.randn(N, d)*sqrt(h)
        X+=sqrt(2)*dB
        t1-=h
        if(viz):
            visualize_dist(X)
        buffer.append(X.clone())
        ts.append(t1)
    return torch.stack(buffer), torch.tensor(ts)

def flatten(layers, ts):
    """
    layers: Tensor of shape (T, N, d)
    ts: Tensor of shape (T,) â€“ time values
    Returns: Tensor of shape (T*N, d+1)
    """
    T, N, d = layers.shape

    # Repeat time values for each sample in a layer
    time_column = ts[:, None].expand(T, N).reshape(-1, 1)  # shape (T*N, 1)

    # Flatten the layer data
    x_flat = layers.reshape(-1, d)  # shape (T*N, d)
    # Concatenate time as the last dimension
    concatenated = torch.cat([x_flat, time_column], dim=1)
    # we take the 1st to last of the time labeled data as x
    # we take the 0th to last-1th of original data as y
    # so that each input maps to the output which is the previous layer's samples
    return concatenated # shape (T*N, d+1)
```

```python
# abstract the procedure of setting up a training into a single function
# h is diffusion time step
def setup_train(N, d, h, t1=1, visualize = False):
    samples = sample(N, d)
    if(visualize):
        visualize_dist(samples.detach().numpy())
    # compute the mean of samples across the sample size dimension
    mean = samples.mean(dim = 0)
    # 0 mean from samples
    samples -= mean
    layers, ts = diffusion_sampler(t1, h, samples)
    if(visualize):
        visualize_dist(layers[-1].detach().numpy())
    xs = flatten(layers, ts)
    variance = layers[-1].var(dim=0)
    return xs, mean, variance, samples
```

```python
def compute_mean_and_cov(arrays):
    """
    Compute the mean and covariance of a list or array of d-dimensional numpy arrays.

    Parameters:
        arrays (np.ndarray): An (n, d) array, where each row is a d-dimensional sample.

    Returns:
        mean (np.ndarray): A (d,) array representing the mean vector.
        cov (np.ndarray): A (d, d) array representing the covariance matrix.
    """
    arrays = np.asarray(arrays)
    # filter out of range
    # arrays = arrays[np.all((arrays > -2.5) & (arrays < 2.5), axis=1)]
    if arrays.ndim != 2:
        raise ValueError("Input should be a 2D array with shape (n_samples, d)")

    mean = np.mean(arrays, axis=0)
    cov = np.cov(arrays, rowvar=False)  # Each row is a sample

    return mean, cov
```

## implementing the network

```python
# We first use fully connected network to approximate S

class FC_Score(nn.Module):
    def  __init__(self, d, hidden_dim = 50):
        super().__init__()
        self.d = d
        # considering up to a smoothness,
        # we estimate the approximation rank to scale
        # with a squareroot of d
        self.hidden_dim = 5*d
        self.l1 = nn.Linear(d,self.hidden_dim)
        self.l2 = nn.Linear(self.hidden_dim, self.hidden_dim)
        self.l3 = nn.Linear(self.hidden_dim, d)
    def modulation(self, t):
        half_dim = self.hidden_dim // 2
        exponents = torch.arange(half_dim).float() / half_dim
        freqs = 10000 ** (-exponents)
        angles = t[:, None] * freqs[None]
        return torch.cat([torch.sin(angles), torch.cos(angles)], dim=-1)
    def s(self, x, ts):
        h = self.l1(x)
        # print(self.modulation(ts).shape)
        h *= self.modulation(ts)
        h = self.l2(h)
        return self.l3(h)

    # x here should absorb the step label as well
    def forward(self, x, h = 1e-3):
        # x.requires_grad_(True)
        score =  self.s(x[:,:-1], x[:,-1])
        # Compute divergence: sum_i d(score_i)/dx_i
        divergence = torch.zeros(x.shape[0], device=x.device)
        for i in range(self.d):
            x_perturbed = x.clone()
            x_perturbed[:, i] += h
            divergence += (self.s(x_perturbed[:,:-1], x_perturbed[:,-1])[:, i] - score[:, i]) / h
        return score, divergence
```

```python
# compute loss on scores
def compute_loss(xs, model):
    n = xs.shape[0]
    d = xs.shape[1] - 1
    ss, divs = model(xs)
    # print(divs.shape)
    return ((ss*ss).sum() + 2*divs.sum())/n/d

def train(N, d, h, t1 = 1, B = 100, model=None, epsilon=1e-4, max_steps=10000):
    xs, mean, var, x0 = setup_train(N, d, h, t1 = t1)
    if(model == None):
        model = FC_Score(d)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    for step in range(max_steps):
        idx = torch.randperm(xs.shape[0])[:B]
        batch = xs[idx]
        loss = compute_loss(batch, model)
        optimizer.zero_grad()
        loss.backward()

        if(step%(min(max_steps//10, 100))==0):
            # Compute total gradient norm
            grad_norm = 0.0
            for p in model.parameters():
                if p.grad is not None:
                    grad_norm += p.grad.norm().item()**2
            grad_norm = grad_norm**0.5

            # Check termination condition
            if grad_norm < epsilon:
                print(f"Stopping early at step {step}, grad norm {grad_norm:.2e}")
                break

            print(f'training {step} step with loss {loss}')
        optimizer.step()

    return model, xs, mean, var
```

```python
score, xs, mean, var = train(1000, 16, 0.01, t1 = 2, B  = 100, max_steps = 5000)
```

```python
# batches a stable diffusion process
# of N samples
def stable_diffusion(score, mean, var, d, N, h, t1):
    y0 = torch.randn(N, d)*var
    t=0
    y = y0
    while(t<=t1+0.00001):
        # construct the d+1 input inference
        x = torch.cat((
            y,
            torch.full((y.shape[0], 1), t, dtype=y.dtype, device=y.device)
        ), dim=1)
        # print(x.shape)
        s,_ = score(x)
        y += 2*s*h + torch.randn(N, d) * sqrt(2*h)
        # if(t<0.5):
            # visualize_dist(y.detach().numpy())
        t += h
    return y+mean
```

## Improved Network

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ScoreNet(nn.Module):
    def __init__(self, d, hidden_dim=128, num_layers=4):
        super().__init__()
        self.d = d
        self.hidden_dim = hidden_dim

        # Time embedding: projects scalar t into a hidden vector
        self.time_embedding = nn.Sequential(
            nn.Linear(1, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, hidden_dim),
        )

        # Input projection
        self.input_proj = nn.Linear(d, hidden_dim)

        # Main body
        self.layers = nn.ModuleList([
            nn.Sequential(
                nn.Linear(hidden_dim, hidden_dim),
                nn.SiLU()
            )
            for _ in range(num_layers)
        ])

        # Output projection to score vector
        self.out_proj = nn.Linear(hidden_dim, d)

    def s(self, x, t):
        # x: [B, d], t: [B, 1]
        t_emb = self.time_embedding(t)        # [B, hidden_dim]
        x_emb = self.input_proj(x)            # [B, hidden_dim]
        h = x_emb + t_emb                     # [B, hidden_dim]

        for layer in self.layers:
            h = layer(h)

        return self.out_proj(h)               # [B, d]

    def forward(self, x):
        # x: [B, d+1]
        x_val = x[:, :-1].clone().detach().requires_grad_(True)
        t_val = x[:, -1:].clone().detach()

        score = self.s(x_val, t_val)  # [B, d]

        # Hutchinson's trick: one backward pass per batch
        noise = torch.randn_like(x_val)
        grad = torch.autograd.grad(
            outputs=(score * noise).sum(),
            inputs=x_val,
            create_graph=True,
            retain_graph=True,
            only_inputs=True
        )[0]

        divergence = (grad * noise).sum(dim=1)  # [B]

        return score, divergence
```

```python
# Enable GPU based training

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# print(device)
def train_gpu(N, d, h, t1=1, B=100, model=None, optimizer=None, epsilon=1e-4, max_steps=10000):
    xs, mean, var, x0 = setup_train(N, d, h, t1=t1)

    # Move data to device
    xs = xs.to(device)
    mean = mean.to(device)
    var = var.to(device)

    if model is None:
        model = FC_Score(d)

    model.to(device)  # Move model to GPU
    if optimizer is None:
      optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    for step in range(max_steps):
        idx = torch.randint(0, xs.shape[0], (B,), device=device)
        batch = xs[idx]  # Already on GPU
        loss = compute_loss(batch, model)

        optimizer.zero_grad()
        loss.backward()

        # print(step)
        if step % (max_steps // 100) == 0:
            grad_norm = 0.0
            for p in model.parameters():
                if p.grad is not None:
                    grad_norm += p.grad.norm().item() ** 2
            grad_norm = grad_norm ** 0.5

            if grad_norm < epsilon or loss < -2:
                print(f"Stopping early at step {step}, grad norm {grad_norm:.2e}")
                break

            print(f'training {step} step with loss {loss.item():.4f}')

        optimizer.step()

    return model, optimizer, xs, mean, var

def stable_diffusion_gpu(score, mean, var, d, N, h, t1):
    device = mean.device  # assumes mean and var are already on the correct device
    score.eval()  # Set model to evaluation mode
    score.to(device)  # Move model to GPU

    y0 = torch.randn(N, d, device=device) * var  # âœ… on GPU
    y = y0
    t = 0.0

    while t <= t1 + 1e-5:
        # construct (y, t) input
        t_tensor = torch.full((y.shape[0], 1), t, dtype=y.dtype, device=device)  # âœ… on GPU
        x = torch.cat((y, t_tensor), dim=1)

        s, _ = score(x)
        # print(s)
        noise = torch.randn(N, d, device=device)  # âœ… noise on GPU
        y += 2 * s * h + noise * sqrt(2 * h)

        t += h

    return y + mean  # mean should be broadcastable and on the same device

```

```python
# prompt: please simplify the run experiment for d logic by reusing previously defined train_gpu

import gc # Import the garbage collection module

def run_experiment_for_d(d, N_values, h = 0.001, inf_h = 0.01, t1=2, use_trained = False, max_steps_per_N=50000):
    """
    Runs experiment (a) comparing sample mean and covariance for different N values.

    Args:
        N_values (list): List of N values to test.
        d (int): Dimension of the samples.
        h (float): Diffusion step size.
        t1 (float): Final diffusion time.
        max_steps_per_N (int): Maximum training steps for each N value.
    """
    mean_errors = []
    cov_errors = []

    # Generate initial data for comparison outside the loop for efficiency
    # Assuming the target distribution doesn't change with N, only sample size does
    # We need a large reference set to compare against
    x0 = sample(100000, d)
    data_mean_ref, data_cov_ref = compute_mean_and_cov(x0.detach().numpy())
    del x0
    print("Reference data generated.")

    for N in N_values:
        print(f"\n--- Running experiment for N = {N}, d = {d} ---")

        # Reuse the train_gpu function
        # We will re-initialize the model and optimizer for each N unless specified otherwise
        conv_score = ScoreNet(d)
        optimizer = None # Let train_gpu create a new optimizer
        # Adjust max_steps based on N if needed, or keep constant
        current_max_steps = max_steps_per_N # Could scale with log(N) or sqrt(N)
        model = conv_score
        if(use_trained):
          conv_score.load_state_dict(torch.load(f'conv_score_{d}_{N}.pth'))
          # run this step to restore initial mean and variance estimations
          xs, mean, var, x0 = setup_train(N, d, h, t1=t1)
          # Move data to device
          del xs
          del x0
          mean = mean.to(device)
          var = var.to(device)
        else:
          # train for two epochs
          # first pretrain on small batches
          model, optimizer, xs, mean, var = train_gpu(
              N//2, d, h, t1=t1,
              model=conv_score, # Pass the potentially new model
              optimizer=optimizer, # Pass None to get a new one
              B= N//10, # First train on small batches
              max_steps=current_max_steps//5
          )

          model, optimizer, xs, mean, var = train_gpu(
              N//2, d, h, t1=t1,
              model=conv_score, # Pass the potentially new model
              optimizer=optimizer, # Pass None to get a new one
              B= N, # Next increase batch size to stablize gradient
              max_steps=current_max_steps
          )

        del xs
        torch.cuda.empty_cache()
        gc.collect()

        torch.save(conv_score.state_dict(), f'conv_score_{d}_{N}.pth')

        print(f"Training finished for N = {N}.")

        # Generate samples from the trained model
        N_generate = 10000 # Generate a fixed large number of samples for consistent evaluation
        print(f"Generating {N_generate} samples from the trained model for N = {N}.")
        ys_batches = []
        batch_size_inf = 1000 # Smaller batch size for inference if memory is an issue

        # batch compute samples
        for i in range(N_generate // batch_size_inf):

          ys = stable_diffusion_gpu(
              model,
              mean.to(device),
              var.to(device),
              d=d,
              N=batch_size_inf,
              h=inf_h, # Inference step size can be different from training h
              t1=t1
          )
          ys_cpu = ys.cpu().detach().numpy()
          ys_batches.append(ys_cpu)

          del ys

          # make sure that there is enough memory in GPU
          torch.cuda.empty_cache()
          gc.collect()

        print(f"Sample generation finished for N = {N}.")
        ys_all = np.concatenate(ys_batches, axis=0)
        print(f"Concatenation finished for N = {N}.")

        # Compute mean and covariance of the generated samples
        trained_mean, trained_cov = compute_mean_and_cov(ys_all)

        del ys_all
        torch.cuda.empty_cache()
        gc.collect()

        # Compute errors relative to the reference data
        mean_error = np.linalg.norm(trained_mean - data_mean_ref)
        cov_error = np.linalg.norm(trained_cov - data_cov_ref) # Frobenius norm

        mean_errors.append(mean_error)
        cov_errors.append(cov_error)

        print(f"N = {N}: Mean Error = {mean_error:.4e}, Covariance Error = {cov_error:.4e}")

    return mean_errors, cov_errors, N_values
```

```python
def plot_errors(N_values, mean_errors, cov_errors, d_val):
    """Plots mean and covariance errors against N on regular and log-log scales."""
    N_values_np = np.array(N_values)
    mean_errors_np = np.array(mean_errors)
    cov_errors_np = np.array(cov_errors)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Regular plot
    axes[0].plot(N_values_np, mean_errors_np, marker='o', label='Mean Error')
    axes[0].plot(N_values_np, cov_errors_np, marker='x', label='Covariance Error')
    axes[0].set_xlabel('Sample Size (N)')
    axes[0].set_ylabel('Error (Frobenius Norm)')
    axes[0].set_title(f'Error vs. Sample Size (d={d_val}) - Regular Scale')
    axes[0].legend()
    axes[0].grid(True)

    # Log-log plot
    axes[1].loglog(N_values_np, mean_errors_np, marker='o', label='Mean Error')
    axes[1].loglog(N_values_np, cov_errors_np, marker='x', label='Covariance Error')
    # Add theoretical O(1/sqrt(N)) decay line
    # Choose a reference point and plot decay relative to it
    if len(N_values) > 1:
      N_ref = N_values_np[1]
      mean_err_ref = mean_errors_np[1]
      cov_err_ref = cov_errors_np[1]
      axes[1].loglog(N_values_np, mean_err_ref * np.sqrt(N_ref / N_values_np), linestyle='--', color='gray', label=r'$O(1/\sqrt{N})$ (Mean Ref)')
      axes[1].loglog(N_values_np, cov_err_ref * np.sqrt(N_ref / N_values_np), linestyle=':', color='purple', label=r'$O(1/\sqrt{N})$ (Cov Ref)')

    axes[1].set_xlabel('Sample Size (N)')
    axes[1].set_ylabel('Error (Frobenius Norm)')
    axes[1].set_title(f'Error vs. Sample Size (d={d_val}) - Log-Log Scale')
    axes[1].legend()
    axes[1].grid(True, which="both", ls="-")

    plt.tight_layout()
    plt.show()




```

The diffusion process of $d=128$ didn't run successfully, due to time steps being too large when generating the samples, causing results to diverge. We will run the SDE with $h=0.0002$. For inference we will rerun the case with the same models, but choosing a smaller time step $h_{inf}=0.001$ for stable diffusion inference. When running backward inference, we batch the 10000 samples into samples of size 1000 in order to prevent running out of GPU memory.

```python

N_values = [400, 1600, 3600, 6400, 10000]
mean_errors_128, cov_errors_128, N_experiment_128 = run_experiment_for_d(128,N_values, inf_h = 0.002, use_trained = False, max_steps_per_N=100000)
%store mean_errors_128
%store cov_errors_128
%store N_experiment_128
```

```python
plot_errors(N_experiment_128, mean_errors_128, cov_errors_128, 128)
```

In the above examples for $d=16, d=32, d= 64$, we see a general trend of decreasing error both in covariance and in mean that is inversely proportional with the square root of sample size, $\frac{1}{\sqrt{N}}$. The match can be observed in the log log plot with a slope of $-\frac{1}{2}$. This was not observed in the case of $d=128$, potentially because having more samples meant more required training epochs before epochs, which exceeded our criteria of 100000 iterations, causing the model to underfit. Usually when we see that the score fitting error fall below -2, it means that the sample features have been fully learned, which was not the case with samples of size greater than 1600 in this case.
